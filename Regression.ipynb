```python
#Simple Linear regression
#used when there is only 1 input feature(independent variable)
#Y=mX+c---> m=lr.coef_
#      ---> c=lr.intercept_   
import pandas as pd
dataset=pd.read_csv(r"D:\Downloads\archive\Salary_dataset.csv")
dataset.head()
```

```python
x=dataset[["YearsExperience"]]
y=dataset["Salary"]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.xlabel("Years of Experience of Employee")
plt.ylabel("Salary of Employee")
plt.grid(True)
sns.scatterplot(x="YearsExperience", y="Salary", data=dataset)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#If we don't set random_state, the data will be split differently every time you run the code.
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#accuracy score
lr.score(x_test,y_test)
```

```python
lr.predict([[1.2]])
#almost correct prediction
```

```python
#to see prediction line
plt.figure(figsize=(7,5))
plt.xlabel("Years of Experience of Employee")
plt.ylabel("Salary of Employee")
plt.grid(True)
sns.scatterplot(x="YearsExperience", y="Salary", data=dataset)
plt.plot(dataset["YearsExperience"],lr.predict(x))
plt.legend(["Dataset","Prediction Line"])
plt.show()
```
