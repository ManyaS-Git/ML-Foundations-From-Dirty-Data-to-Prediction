```python
#POLYNOMIAL REGRESSION
#y = b0 + b1*x + b2*x^2 + b3*x^3 + ... + bn*x^n
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv(r"D:\Downloads\archive\salary.csv")
dataset.head()
```

```python
print("Columns in dataset:", dataset.columns.tolist())
```

```python
sns.scatterplot(x="Level ", y="Salary", data=dataset)
plt.show()
'''
EXPLANATION OF THE MODELS:
1. Simple Linear Regression (y = b0 + b1x): 
   Fails here because it assumes a straight-line relationship. 
   The red "error" in the diagram shows the high distance between 
   the straight line and the curved data (Underfitting).

2. Polynomial Regression (y = b0 + b1x + b2x^2): 
   Solves the error by allowing the regression line to curve.
   This "bends" the model to fit the non-linear trend of the 
   actual data points, resulting in much higher accuracy.
'''
```

```python
x=dataset[["Level "]]
y=dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pr=PolynomialFeatures(degree=4)
pr.fit(x)
x = pr.transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
pred=lr.predict(x)
sns.scatterplot(x="Level ", y="Salary",data=dataset)
plt.plot(dataset["Level "],pred)
plt.show()
```
