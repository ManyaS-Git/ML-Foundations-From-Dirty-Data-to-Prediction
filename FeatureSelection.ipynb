```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset=pd.read_csv(r"C:\Users\MANYA SHARMA\Machine Learning/transformed_retail_sales.csv")
dataset.head()
```

```python
#Feature Select Tecinques
#selecting the columns which influence the outcome
#feature is an attribute that has an impact on a problem or is useful for the problem, and choosing the important
#features for thhe model is known as feature selection
```

```python
#Forward Elimination
#check the Ml model on eachh feature and check accuracy
#combine 2-2 features with the highest accuracy and again check accurary
#accuracy must increase and this way we keep o moving ahead

#Backward Elimination
#calculate the accuracy of all features and eleminate the highest one
#make groups and carry on the same
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
X = dataset.drop(columns=['Transaction ID', 'Customer ID', 'Transaction Date', 'Total Spent', 'Total Spent Transformed'])
y = dataset["Total Spent Transformed"]
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
sfs=SequentialFeatureSelector(lr,k_features=3,forward=True)
#lr checks how good the features work together
#k_features is the number of useful columns
sfs.fit(X,y)
sfs.feature_names
```

```python
sfs.k_feature_names_
```

```python
sfs.k_score_ #accuracy
```

```python
selected_columns = list(sfs.k_feature_names_)
X_final = X[selected_columns]

print("Feature selection complete. Proceeding with features:", selected_columns)
```
