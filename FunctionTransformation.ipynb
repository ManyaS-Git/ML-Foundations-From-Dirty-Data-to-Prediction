```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
dataset=pd.read_csv(r"C:\Users\MANYA SHARMA\Data Cleaning\cleaned_retail_sales.csv")
dataset.head()
```

```python
sns.distplot(dataset["Total Spent"])
plt.title("Distribution Before Transformation")
plt.show()
```

```python
#FUNCTION TRANSFORMATION
#transform your data into a more "Normal" (Gaussian) distribution
from sklearn.preprocessing import FunctionTransformer
ft=FunctionTransformer(func=np.log1p)
# Applying Log Transformation to 'Total Spent'
# We add 1 (log1p) to avoid errors if there are zero values
dataset["Total Spent Transformed"]=ft.fit_transform(dataset[["Total Spent"]])
```

```python
sns.distplot(dataset["Total Spent Transformed"])
plt.title("Distribution After Transformation")
plt.show()
```

```python
#Plotting before and after side by side
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
sns.distplot(dataset["Total Spent"])
plt.title("Before")

plt.subplot(1,2,1)
sns.distplot(dataset["Total Spent Transformed"])
plt.title("After")
plt.tight_layout()
plt.show()
```

```python
#When to use Function Transformer
#when your numerical data does not follow a Normal Distribution (the "Bell Curve")
dataset.to_csv("transformed_retail_sales.csv", index=False)
print("File saved successfully!")
```
